## Legends, Axes label, Grid

```python
days = np.arange(1,8)
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]

```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Data')

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

plt.legend()                    #to show the legend            #plt.legend(loc='best')
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Data')

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
 
    
plt.legend(loc="upper right", shadow='True', fontsize='small')  


plt.grid()                      # to make grid visible
```

## Plotting two bar graphs 

```python
company = ['amazon','microsoft','google','fb']
revenue = [90, 135, 95, 30]
profit = [40, 10, 32, 12]
xpos = np.arange(len(company))
plt.xticks(ypos, company)


plt.xlabel('Company')
plt.ylabel('Million dollars')
plt.title('US Tech Stocks')





plt.bar(xpos, revenue, label='Revenue')
plt.bar(xpos, profit, label='Profit')

plt.legend()
```

### To separate out each column:

```python
# this now gives two separate bars

company = ['amazon','microsoft','google','fb']
revenue = [90, 135, 95, 30]
profit = [40, 10, 32, 12]
xpos = np.arange(len(company))
plt.xticks(ypos, company)


plt.xlabel('Company')
plt.ylabel('Million dollars')
plt.title('US Tech Stocks')



plt.bar(xpos - 0.2, revenue, label='Revenue')      #shift the bars by xpos+.2(thats why used numpy array for xpos)
plt.bar(xpos + 0.2, profit, label='Profit')        #on adding + 0.2 will add to all elements in the array

plt.legend()
```

```python
#bars with adjusted widths


company = ['amazon','microsoft','google','fb']
revenue = [90, 135, 95, 30]
profit = [40, 10, 32, 12]
xpos = np.arange(len(company))
plt.xticks(ypos, company)


plt.xlabel('Company')
plt.ylabel('Million dollars')
plt.title('US Tech Stocks')



plt.bar(xpos - 0.2, revenue, label='Revenue', width = 0.4)      
plt.bar(xpos + 0.2, profit, label='Profit',  width = 0.4)       

plt.legend()
```

# Histograms

```python
blood_sugar = [113,85,93,92,120,117,102,97,99,107,88,77]

plt.hist(blood_sugar)              # by default it plots 10 bins(buckets)(ranges) for you
```

```python
# we can adjust bin size, and also rwidth which is relative width of a bar wrt the bin

blood_sugar = [113,85,93,92,120,117,102,97,99,107,88,77]

plt.hist(blood_sugar, bins=3, rwidth=0.95)              
```

```python
# we can also specify the bins' ranges explicitly


blood_sugar = [113,85,93,92,120,117,102,97,99,107,88,147]

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='green')
```

```python
# to make side by side histograms, provide the arguments as a list


blood_sugar_men = [113,85,93,92,120,117,102,97,99,107,88,147]
blood_sugar_women = [120,130,89,98,95,143,82,102,118,135,99,107]


plt.xlabel('Sugar range')
plt.ylabel('Frequency (no. of people)')
plt.title('Blood sugar analysis')


plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,160], rwidth=0.90, color=['green', 'orange'],
        label = ['mean','women'])


plt.legend()                       # enable legend after you mention labels in the plt.hist()
```

```python
# orientations can be specified as horizontal


blood_sugar_men = [113,85,93,92,120,117,102,97,99,107,88,147]
blood_sugar_women = [120,130,89,98,95,143,82,102,118,135,99,107]


plt.xlabel('Sugar range')
plt.ylabel('Frequency (no. of people)')
plt.title('Blood sugar analysis')


plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,160], rwidth=0.90, color=['green', 'orange'],
        label = ['mean','women'], orientation='horizontal')


plt.legend()                       
```

# Pie Chart

```python
# pie chart for expenses

exp_values = [1400, 600, 300, 410, 820]
exp_labels = ['Home Rent', 'Food', 'Car', 'Phone', 'Others']

plt.pie(exp_values, labels=exp_labels)

plt.legend()
```

```python
exp_values = [1400, 600, 300, 410, 820]
exp_labels = ['Home Rent', 'Food', 'Car', 'Phone', 'Others']


plt.axis('equal')                                                # to ensure perfect circle
plt.pie(exp_values, labels=exp_labels, radius=1.8)               # by default radius is 1


plt.show()
```

```python
# to specify percentages in the pie chart (using autopct) :


exp_values = [1400, 600, 300, 410, 820]
exp_labels = ['Home Rent', 'Food', 'Car', 'Phone', 'Others']


plt.axis('equal')                                               
plt.pie(exp_values, labels=exp_labels, radius=1.8, autopct = '%0.2f%%')                 

plt.show()


# autopct = '%0.0f%%' for no decimal
# autopct = '%0.2f%%' for upto two decimals
```

```python
# to bring certain pieces of the pie out : explode which takes list of arguements

exp_values = [1400, 600, 300, 410, 820]
exp_labels = ['Home Rent', 'Food', 'Car', 'Phone', 'Others']


plt.axis('equal')                                               
plt.pie(exp_values, labels=exp_labels, radius=1.8, autopct = '%0.2f%%', explode=[0,0.2,0,0,0])                 

plt.show()
```

```python
exp_values = [1400, 600, 300, 410, 820]
exp_labels = ['Home Rent', 'Food', 'Car', 'Phone', 'Others']


plt.axis('equal')                                               
plt.pie(exp_values, labels=exp_labels, radius=1.8, autopct = '%0.2f%%', explode=[0,0.2,0,0.4,0])                 

plt.show()
```
